```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('images/bg.jpg')
k = 0
X = np.random.randint(0,1,(100,2))
Y = np.random.randint(0,1,(100,2))
squareSide = 30
for j in range(2, 600, 50):
    for i in range(1, 350, 40):
        if(k < 100):
            X[k] = [i + squareSide/2, j+ squareSide/2]
            cv2.rectangle(img, (i, j), (i + squareSide, j + squareSide), (0, 255, 0), 2)
            k+=1

k = 0
for j in range(int(squareSide / 2) + 2, 600, 50):
    for i in range(450, 800, 40):
        if(k < 100):
            Y[k] = [i, j]
            p1 = (i - int(squareSide/2), j - int(squareSide/2))
            p2 = (i - int(squareSide/2), j + int(squareSide/2))
            p3 = (i + int(squareSide/2), j + int(squareSide/2))
            cv2.line(img, p1, p2, (255, 0, 0), 2)
            cv2.line(img, p2, p3, (255, 0, 0), 2)
            cv2.line(img, p1, p3, (255, 0, 0), 2)
            k+=1

cv2.imwrite('images/image1.jpg', img)

Z = np.vstack((X,Y))
Z = np.float32(Z)
# define criteria and apply kmeans()
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 10, 1.0)
ret,label,center=cv2.kmeans(Z,2,None,criteria,10,cv2.KMEANS_PP_CENTERS)
# Now separate the data, Note the flatten()
A = Z[label.ravel()==0]
B = Z[label.ravel()==1]
# Plot the data
plt.scatter(A[:,0],A[:,1])
plt.scatter(B[:,0],B[:,1],c = 'r')
plt.scatter(center[:,0],center[:,1],
s = 80,c = 'y', marker = 's')
plt.xlabel('Height'),plt.ylabel('Weight')
plt.show()
```
