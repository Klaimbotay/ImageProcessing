```python
import random
import copy

#генерируем список mas, состоящий из 100 элементов, заполняем случайными числами [0, 100], округленными до 2х знаков после запятой
mas = [round(random.uniform(0, 100), 2) for i in range(100)]

#все числа списка mas сравниваем с 10 и 90, если число меньше 10, то делаем его равным 10, если больше 90, то равным 90
i = 0
while i < 100:
    if mas[i] < 10:
        mas[i] = 10
    elif mas[i] > 90:
        mas[i] = 90
    i += 1

#рекурсивная функция нахождения суммы всех элементов списка mas
def listsum(numList):
   if len(numList) == 1:
        return numList[0]
   else:
        return numList[0] + listsum(numList[1:])

#находим среднее значение из списка mas
midSum = listsum(mas) / 100


print(midSum, max(mas), min(mas))

sum(mas) / len(mas)
```

```python
#округляем все числа из списка mas до целых
i = 0
while i < 100:
    mas[i] = round(mas[i])
    i += 1

#преобразуем список mas в множество set1
set1 = set(mas)

#создаем set2 - полную копию мн-ва set1
set2 = copy.deepcopy(set1)

#удаляем первые три элемента из мн-ва set1
set1.pop()
set1.pop()
set1.pop()

#выводим разность мн-в set2 и set1
print(set2.difference(set1))
```

```python
#создаем словарь autoshop1
autoshop1 = {
    "Toyota" : 2400000,
    "Bmw" : 3000000,
    "Mercedes-Benz" : 4000000,
    "Volkswagen" : 2700000,
    "Audi" : 3550000,
    "Skoda" : 1500000,
    "Nissan" : 2100000,
    "Mazda" : 2500000,
    "Subaru" : 2150000,
    "Mitsubishi" : 2050000,   
}

#создаем autoshop2 - полную копию autoshop1 
autoshop2 = copy.deepcopy(autoshop1)

#находим минимальное значение в autoshop2
min_value = min(autoshop2.values())

#проходим по словарю, запоминаем какому ключу принадлежит минимальное значение и удаляем элемент словаря
for key in autoshop2:
    if autoshop2[key] == min_value:
        e = key
del autoshop2[e]

for key in autoshop2:
    autoshop2[key] = random.randint(2000000, 4000000)

#добавляем элемент в словарь autoshop2
autoshop2.update({"Kia": 2158302})

#функция поиска автомобиля в двух словарях и выбор выгодного варианта, выводит название магазина и цену
def podbor(auto):
    if auto in autoshop1 and auto in autoshop2:
        if autoshop1[auto] <= autoshop2[auto]:
            print("autoshop1 : ", autoshop1[auto])
            return
        else:
            print("autoshop2 : ", autoshop2[auto])
            return
    elif auto in autoshop1:
        print("autoshop1 : ", autoshop1[auto])
        return
    elif auto in autoshop2:
        print("autoshop2 : ", autoshop2[auto])
        return
    else:
        print("This car dosn't exist")
        return
    
print(autoshop1,'\n')
print(autoshop2,'\n')

podbor(input())
```
