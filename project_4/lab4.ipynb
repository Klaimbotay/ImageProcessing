```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

image = cv2.imread('images/image.jpg', 0)

plt.hist(image.ravel(),256,[0,256])

plt.show()
```

```python
img = cv2.imread('images/image.jpg')
```

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

mask1 = np.zeros(img.shape[:2], np.uint8)
mask2 = np.zeros(img.shape[:2], np.uint8)
mask3 = np.zeros(img.shape[:2], np.uint8)

mask1[:, 0:300] = 255
mask2[0:300, 301:799] = 255
mask3[301:479, 301:799] = 255

masked_img1 = cv2.bitwise_and(img,img,mask = mask1)
masked_img2 = cv2.bitwise_and(img,img,mask = mask2)
masked_img3 = cv2.bitwise_and(img,img,mask = mask3)

plt.subplot(221)
plt.imshow(masked_img1,'gray')
plt.subplot(222)
color = ('b','g','r')
for i,col in enumerate(color):
    histr = cv2.calcHist([img],[i],mask1,[256],[0,256])
    plt.plot(histr,color = col)
    plt.xlim([0,256])
plt.show()

plt.subplot(223)
plt.imshow(masked_img2,'gray')
plt.subplot(224)
for i,col in enumerate(color):
    histr = cv2.calcHist([img],[i],mask2,[256],[0,256])
    plt.plot(histr,color = col)
    plt.xlim([0,256])
plt.show()

plt.subplot(223)
plt.imshow(masked_img3,'gray')
plt.subplot(224)
for i,col in enumerate(color):
    histr = cv2.calcHist([img],[i],mask3,[256],[0,256])
    plt.plot(histr,color = col)
    plt.xlim([0,256])
plt.show()
```

```python
img = cv2.imread('images/image_3.jpg')
img_yuv = cv2.cvtColor(img, cv2.COLOR_BGR2YUV)

img_yuv[:,:,0] = cv2.equalizeHist(img_yuv[:,:,0])

img_output = cv2.cvtColor(img_yuv, cv2.COLOR_YUV2BGR)

cv2.imwrite('images/image_3new.jpg', img_output)
```

```python
img = cv2.imread('images/image_sudoku.jpg', 0)

ret,thresh1 = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)
ret,thresh2 = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY_INV)
ret,thresh3 = cv2.threshold(img, 127, 255, cv2.THRESH_TRUNC)
ret,thresh4 = cv2.threshold(img, 127, 255, cv2.THRESH_TOZERO)
ret,thresh5 = cv2.threshold(img, 127, 255, cv2.THRESH_TOZERO_INV)

titles = ['OriginalImage','BINARY','BINARY_INV','TRUNC','TOZERO','TOZERO_INV']

cv2.imwrite('images/OriginalImage.jpg', img)
cv2.imwrite('images/BINARY.jpg', thresh1)
cv2.imwrite('images/BINARY_INV.jpg', thresh2)
cv2.imwrite('images/TRUNC.jpg', thresh3)
cv2.imwrite('images/TOZERO.jpg', thresh4)
cv2.imwrite('images/TOZERO_INV.jpg', thresh5)
```

```python
img = cv2.imread('images/image_sudoku.jpg',0)
th1 = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11, 2)
th2 = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 5, 2)
th3 = cv2.adaptiveThreshold(img, 150, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11, 2)
th4 = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY_INV, 11, 2)
th5 = cv2.adaptiveThreshold(img, 150, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY_INV, 11, 2)
#adaptiveThreshold(src, maxValue, adaptiveMethod, thresholdType,blockSize, C[, dst])

cv2.imwrite('images/adaptive1.jpg',th1)
cv2.imwrite('images/adaptive2.jpg',th2)
cv2.imwrite('images/adaptive3.jpg',th3)
cv2.imwrite('images/adaptive4.jpg',th4)
cv2.imwrite('images/adaptive5.jpg',th5)
```
